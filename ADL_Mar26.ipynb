### standardize names

normalizes school names (lowercase, removing text within parentheses, replacing hyphens with spaces, removing other common punctuation, and stripping extra spaces)


```python
import re

def clean_school_name(name):
    # lowercase
    name = str(name).lower()

    # priority matching
    if "perimeter" in name:
        return "perimeter"

    # get rid of parentheses and their contents
    name = re.sub(r'\(.*?\)', '', name)
    # delete hyphens
    name = name.replace('-', ' ')
    # remove other common punctuation, replace & with and
    name = re.sub(r'[^a-z0-9\s&]', '', name)
    name = name.replace('&', 'and')
    # remove 'college', 'university', 'of', 'nursing', 'son', 'undergraduate', 'school of nursing', and 'campus'
    name = name.replace('college', '').replace('university', '').replace('of', '').replace('nursing', '').replace('son', '').replace('undergraduate', '').replace('school of nursing', '').replace('campus', '')
    # replace 'ga' with 'georgia'
    name = re.sub(r'\bga\b', 'georgia', name)
    # replace multiple spaces with a single, strip leading/trailing spaces
    name = re.sub(r'\s+', ' ', name)
    name = name.strip()

    return name
```

```python
affliliated['clean_school'] = affliliated['School'].apply(clean_school_name)
partnered['clean_school'] = partnered['Facility'].apply(clean_school_name)
nclex['clean_school'] = nclex['Program_Name'].apply(clean_school_name)
vizient['clean_school'] = vizient['education'].apply(clean_school_name)
```

### fuzzy matching

note:

for Auburn University - School of Nursing / Auburn University - College of Nursing (Undergraduate)
and Emory University - SON / Emory University - SON Pre-licensure

rotations data is more granular  so will map nclex/affiliated/partnered onto both just note that those are school level as opposed to program level

```python
from fuzzywuzzy import process

unique_rotations_schools = rotations['school_id'].unique()

# extract unique school lists
other_schools_nclex = nclex['clean_school'].unique()
other_schools_affiliated = affliliated['clean_school'].unique()
other_schools_partnered = partnered['clean_school'].unique()

# weak tokens to ignore for matching
weak_tokens = {"georgia", "state", "technical"}

def has_strong_overlap(a, b, ignore_weak=False):
    if ignore_weak:
        return True

    a_tokens = set(a.lower().split()) - weak_tokens
    b_tokens = set(b.lower().split()) - weak_tokens

    # if both strings have no strong tokens then allow match
    if not a_tokens and not b_tokens:
        return True

    return len(a_tokens & b_tokens) > 0

# schools with more specific separations
school_groups = {
    "auburn": [r for r in unique_rotations_schools if "auburn" in r.lower()],
    "emory": [r for r in unique_rotations_schools if "emory" in r.lower()],
    "herzing": [r for r in unique_rotations_schools if "herzing" in r.lower()]
}

fuzzy_matches = {}

# ignore_weak = True for grouped schools
for school_id, rotation_programs in school_groups.items():
    representative = "auburn" if school_id == "auburn" else "emory" if school_id == "emory" else "herzing"

    # raw fuzzy matches
    raw_nclex = process.extractBests(representative, other_schools_nclex, score_cutoff=70, limit=10)
    raw_affiliated = process.extractBests(representative, other_schools_affiliated, score_cutoff=70, limit=10)
    raw_partnered = process.extractBests(representative, other_schools_partnered, score_cutoff=70, limit=10)

    # filter matches using has_strong_overlap with ignore_weak=True
    matches_nclex = [m for m in raw_nclex if has_strong_overlap(representative, m[0], ignore_weak=True)][:3]
    matches_affiliated = [m for m in raw_affiliated if has_strong_overlap(representative, m[0], ignore_weak=True)][:3]
    matches_partnered = [m for m in raw_partnered if has_strong_overlap(representative, m[0], ignore_weak=True)][:3]

    for program in rotation_programs:
        fuzzy_matches[program] = {
            'nclex_matches': matches_nclex,
            'affiliated_matches': matches_affiliated,
            'partnered_matches': matches_partnered
        }

# all other schools
other_rotation_programs = [r for r in unique_rotations_schools
                           if all(school not in r.lower() for school in school_groups.keys())]

for rotation_school in other_rotation_programs:
    raw_nclex = process.extractBests(rotation_school, other_schools_nclex, score_cutoff=90, limit=10)
    raw_affiliated = process.extractBests(rotation_school, other_schools_affiliated, score_cutoff=90, limit=10)
    raw_partnered = process.extractBests(rotation_school, other_schools_partnered, score_cutoff=90, limit=10)

    # use weak-tokens
    matches_nclex = [m for m in raw_nclex if has_strong_overlap(rotation_school, m[0])][:3]
    matches_affiliated = [m for m in raw_affiliated if has_strong_overlap(rotation_school, m[0])][:3]
    matches_partnered = [m for m in raw_partnered if has_strong_overlap(rotation_school, m[0])][:3]

    fuzzy_matches[rotation_school] = {
        'nclex_matches': matches_nclex,
        'affiliated_matches': matches_affiliated,
        'partnered_matches': matches_partnered
    }

# print for review - see https://docs.google.com/spreadsheets/d/1BxyT4GXb7WU-O5VM7NTVnJYR3h_mPze9vCc0WCnaB3s/edit?usp=sharing
for school, matches in fuzzy_matches.items():
    print(f"\nRotation School: {school}")
    print(f"  NCLEX Matches: {matches['nclex_matches']}")
    print(f"  Affiliated Matches: {matches['affiliated_matches']}")
    print(f"  Partnered Matches: {matches['partnered_matches']}")
```

## school level dataset

aggregate rotation stats


*   school_id
*   school
*   school_deg_cat
*   prop completed / withdrawn / denied
*   prop precepted rotation (higher is better)
*   prop filled: sum of num_students / num_avail
*   total hours: sum of total hours per person * num_avail for completed only
*   is_affiliated
*   partner_tier





```python
# school_level final
school_level_final = rotation_level_final.copy()

# total assigned hours for each rotation entry
school_level_final['total_assigned_hours'] = school_level_final['total_hours_per_person'] * school_level_final['num_avail']

# total hours ROTATED (status_cat = 1)
school_level_final['completed_assigned_hours'] = school_level_final.apply(
    lambda row: row['total_assigned_hours'] if row['status_cat'] == 1 else 0,
    axis=1
)

school_level_final = school_level_final.groupby(['school_id', 'school', 'school_deg_cat']).agg(
    prop_completed=('status_cat', lambda x: (x == 1).mean()),
    prop_denied=('status_cat', lambda x: (x == 2).mean()),
    prop_withdrawn=('status_cat', lambda x: (x == 3).mean()),
    prop_precepted_rotation=('experience_cat', lambda x: x.mean()),
    total_students_assigned=('num_students', 'sum'),
    total_available_slots=('num_avail', 'sum'),
    total_hours_rotated=('completed_assigned_hours', 'sum'),
    is_affiliated=('is_affiliated', 'first'),
    partner_tier=('partner_tier', 'first')
).reset_index()

school_level_final['prop_filled'] = (school_level_final['total_students_assigned'] / school_level_final['total_available_slots']).fillna(0)

school_level_final = school_level_final[[
    'school_id', 'school', 'school_deg_cat', 'prop_completed',
    'prop_denied', 'prop_withdrawn', 'prop_precepted_rotation',
    'prop_filled', 'total_hours_rotated', 'is_affiliated', 'partner_tier'
]]
```

adding nclex data:

each school will have multiple entries by year and/or type

extra column at end with average pass rate of all years per type

```python
# match mapping
nclex_school_matches_df = pd.DataFrame([
    {'school_id': school_id, 'matched_nclex_clean_school': school}
    for school_id, matches in fuzzy_matches.items()
    for school, _ in (matches.get('nclex_matches') or [])
])

nclex_cols = ['Type', 'Year', 'TotalDelivered', 'TotalPassed', 'TotalFailed', 'PassRate']

if nclex_school_matches_df.empty:
    for col in ['Program_Name'] + nclex_cols + ['average_nclex_pass_rate']:
        school_level_final[col] = pd.NA
else:
    # expand school_level_final rows for each matched school
    school_level_final = school_level_final.merge(
        nclex_school_matches_df, on='school_id', how='left'
    )

    # join in all NCLEX rows (one per Type+Year combo), duplicating as needed
    school_level_final = school_level_final.merge(
        nclex[['clean_school', 'Program_Name'] + nclex_cols],
        left_on='matched_nclex_clean_school',
        right_on='clean_school',
        how='left'
    ).drop(columns=['matched_nclex_clean_school', 'clean_school'], errors='ignore')

    # avg pass rate across all years, per school and type
    school_level_final['average_nclex_pass_rate'] = (
        school_level_final.groupby(['school_id', 'Type'])['PassRate'].transform('mean')
    )
```
